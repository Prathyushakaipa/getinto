```
import pandas
df=pandas.read_csv('/content/drive/My Drive/dataset/bitcoin.csv')
import numpy as np
n1=np.array(df['Open'][:90])
n2=np.array(df['Close'][:90])
n3=n1-n2
date=np.array(df['Date'][:90])
import matplotlib.pyplot as plt
plt.plot(date,n3,'bo-')
plt.xlabel('Date')
plt.ylabel('Gross profit or loss')
plt.show()
```
